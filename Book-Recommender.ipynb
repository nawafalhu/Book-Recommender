```python
rating_book = pd.merge(rating, book, on='ISBN')
```

```python
rating_book
```

```python
df = pd.merge(rating_book, user, on='User-ID')
```

```python
df.head()
```

```python
column = ['User-ID', 'ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Book-Rating', 'Age', 'Location', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L']
df = df[column]
df.head()
```

```python
df.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1, inplace=True)
```

```python
df
```

```python
num_rating = df['Book-Title'].value_counts()
num_rating.head(10)
```

```python
rating = df.groupby('Book-Title').agg(
    Rating=('Book-Rating', 'mean'),  # Calculate mean rating
    Num_of_ratings=('Book-Title', 'count')  # Count the number of ratings
).sort_values(by='Num_of_ratings', ascending=False)

rating.head(20)
```

```python
rating.tail(10)
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.countplot(x=df['Book-Rating']);
plt.title("Book Rating")
plt.xlabel("Rating")
plt.ylabel("Count")
```
